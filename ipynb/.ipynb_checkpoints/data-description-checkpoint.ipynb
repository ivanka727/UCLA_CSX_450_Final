```r
install.packages("pryr", repos = "http://cran.us.r-project.org")
```

```r
library(pryr)
```

```r
abaloneURL <- "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
```

```r
abalone <- read.csv(url(abaloneURL), sep = ",", header = TRUE)
```

```r
object_size(abalone)
```

```r
dim(abalone)
head(abalone)
names(abalone)[1]<-paste("Sex")
names(abalone)[2]<-paste("Length")
names(abalone)[3]<-paste("Diameter")
names(abalone)[4]<-paste("Height")
names(abalone)[5]<-paste("Whole weight")
names(abalone)[6]<-paste("Shucked weight")
names(abalone)[7]<-paste("Viscera weight")
names(abalone)[8]<-paste("Shell weight")
names(abalone)[9]<-paste("Rings")
```

```r
str(abalone)
```

```r
summary(abalone)
```

```r
pairs(abalone[2:8], pch = 21,bg = c("red", "green", "blue", "pink", "purple", "brown", "orange")[unclass(abalone$Sex)])
```

```r
plot(abalone$Length, abalone$Rings, pch = 21,bg = c("red", "green", "blue")[unclass(abalone$Sex)])
```

```r
ggplot(abalone) + aes(Length, Rings, color = Sex) + geom_point() + labs(x = "Length", 
    y = "Rings", title = "Length & Rings") + facet_grid(. ~ 
    Sex, labeller = label_both) + stat_smooth(method = "lm", se = FALSE) + theme(legend.position = "none")
```

```r
library(repr)
options(repr.plot.width=12,repr.plot.length=4)
hist(abalone$Rings, freq = FALSE)
lines(density(abalone$Rings))
```

```r
library(repr)
options(repr.plot.width=12,repr.plot.length=4)
hist(abalone$Length, freq = FALSE)
lines(density(abalone$Length))
```

```r
library(repr)
options(repr.plot.width=12,repr.plot.length=4)
hist(abalone$"Whole weight", freq = FALSE)
lines(density(abalone$"Whole weight"))
```
